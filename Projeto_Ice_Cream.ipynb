## Projeto Ice Cream

```python
# importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('C:\\Users\\david\\Date_Science\\Cientista de Dados Python para Iniciantes\\dataset\\IceCreamData.csv')
df
```

```python
# Conversão de fahrenheit para Celsius
celsius = (df['Temperature'] - 32) *5/9
df['Temperature'] = celsius
df
```

```python
result = df.sort_values('Temperature', ascending=False)
result
```

```python
plt.title ('Ice Cream Truck') # tilulo do grafico

plt.plot(result['Temperature'], result['Revenue'], 'r.--', label = 'US$') # 'label' legenda do grafico 


plt.xticks(df['Temperature'][::2]) # visualizar valores a cada dois anos
plt.xlabel ('Temperatura (Fahrenheit)') 
plt.ylabel ('Sales (US$)') #titulo (descrição lateral)

plt.legend() # legenda  
plt.savefig('Results/Ice_Cream.pdf')
plt.show()
```
