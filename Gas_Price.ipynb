 # Importando o Dataset

```python
# importando as bibliotecas necess√°rias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('C:\\Users\\david\\Date_Science\\Cientista de Dados Python para Iniciantes\dataset\\gas_prices.csv')
gas
```

# Criando o primeiro grafico

```python
plt.plot(gas['Year'], gas['Australia'])
plt.plot(gas['Year'], gas['Italy'])
plt.plot(gas['Year'], gas['USA'])

plt.show()
```

# Adicionando Legenda e Ticks

```python
plt.title ('Valor do Conbustivel em US$') # tilulo do grafico

plt.plot(gas['Year'], gas['Australia'], 'r.-', label = 'Australia') # 'label' legenda do grafico 
plt.plot(gas['Year'], gas['Italy'], 'y.-', label = 'Italy') # 'y.-' Y cor, .- ponto na linha 
plt.plot(gas['Year'], gas['USA'], 'b.-', label = 'USA')
plt.plot(gas['Year'], gas['Japan'], 'g.-', label = 'Japan')

plt.xticks(gas['Year'][::2]) # visulizar valores a cada dois anos

plt.xlabel ('Years') 
plt.ylabel ('US$') #titulo (descri√ß√£o lateral)

plt.legend() # legenda  
plt.show()
```

# Adicionando com o for loop

```python
plt.title ('Valor do Conbustivel em US$') # tilulo do grafico

list_country = ['Australia', 'USA', 'Italy', 'Mexico', 'France']
    
for country in gas:
    if country in list_country:
        plt.plot(gas['Year'], gas [country], label= [country], marker = '.')

plt.xticks(gas['Year'][::2]) # visulizar valores a cada dois anos

plt.xlabel ('Years') 
plt.ylabel ('US$') #titulo (descri√ß√£o lateral)

plt.legend() # legenda  
plt.savefig('Results/country.png') # salvando grafico ( obs.'savefig', sempre antes show )
plt.savefig('Results/country.pdf') # salvando grafico
plt.show()

```
