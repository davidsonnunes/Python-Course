```python
# importando as bibliotecas necessárias
import pandas as pd
import re # 
```

```python
# diretório onde estão armazenado arquivo da pesquisa
df = pd.read_csv('C:\\Users\\david\\Date_Science\\Cientista de Dados Python para Iniciantes\\dataset\\house.csv')
```

```python
df
```

```python
# Filtrando dados com condições
# Filtro ('Type' casa) de ('Rooms' quartos) com valor ('Price' <= 500000)
filter1 = df.loc[(df['Rooms'] == 3) & (df['Type'] == 'h') & (df['Price'] <= 500000)]
filter1
```

```python
# Filtrando com Regular Expression
# Filto por coluna 'Address', por nomes dentro da coluna maius. ou menusc.
filter2 = df.loc[df['Address'].str.contains('Turner st|Turner Rd', flags=re.I)]
filter2
```

```python
# Filtrando com Regx e Conditions
# filtro na coluna 'Address', que começa com numero aparti 59 É valor  'Price' menor ou igual a 500,000
filter3 = df.loc[df['Address'].str.contains('^59', flags=re.I) & (df['Price'] <= 500000)]
filter3
```

```python
# Modificando dados na DataFrame
df.loc[df['SellerG'] == 'Nelson', 'SellerG'] = 'Ronald'
df
```

```python
# Alterando dado em uma coluna 
df.loc[df['SellerG'] == 'Ronald', ['Method']] = 'Pending'
df
```

```python
# Alterando dado em mais de umas coluna 
df.loc[df['SellerG'] == 'Ronald', ['Method', 'Type']] = 'Pending'
df
```

```python
# Agrupando dados com o Groupby
# media de vendedores
df.groupby(['SellerG']).mean()
```

```python
# Média de vendedores
df.groupby(['SellerG']).sum().sort_values('Price', ascending=False).head(10)
```
